```
import autoplot as ap
```

```
org= ap.javaaddpath( 'http://autoplot.org/latest/autoplot.jar'  )
```

```
apds = org.autoplot.idlsupport.APDataSet()
```

```
if apds.loadDataSet( 'http://www.wavsource.com/snds_2018-06-03_5106726768923853/animals/horse.wav' ): print( 'fail to read' )
```

```
print( apds.toString() ) 
```

```
epoch= ap.to_ndarray( apds, 'ds_1' )
horse= ap.to_ndarray( apds, 'ds_0' )
```

```
from matplotlib import pyplot as plt
plt.plot(epoch,horse)
plt.show()
```

```
ep1= epoch / ( 1+1/12. )
ep2= epoch / ( 1+2/12. )
ep3= epoch / ( 1+3/12. )
```
